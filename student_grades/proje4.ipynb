```python
#Gerekli kütüphaneleri import edelim.

import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.utils import shuffle
```

```python
# CSV dosyasını URL üzerinden yükleme
url = "https://raw.githubusercontent.com/amankharwal/Website-data/master/student-mat.csv"
data = pd.read_csv(url)

# Veri setinin ilk 5 satırını görüntüleme
data.head()
```

 Bu veri setinde G1, ilk dönemin notlarını, G2, ikinci dönemin notlarını ve G3, final notlarını temsil eder. 
 Şimdi verileri hazırlayalım ve öğrencilerin final notlarını nasıl tahmin edebileceğimizi görelim.

```python
import numpy as np
import pandas as pd

#final notu tahmini
data = data[["G1", "G2", "G3", "studytime", "failures", "absences"]]
predict = "G3"

x = np.array(data.drop([predict], axis=1))  # 'G3' sütununu düşürmek için axis=1
y = np.array(data[predict])

# Eğitim ve test verilerine ayırma
from sklearn.model_selection import train_test_split

#veri setini %20 test ve %80 eğitim olarak böldük.
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
linear_regression = LinearRegression()
linear_regression.fit(X_train, y_train)
accuracy = linear_regression.score(X_test, y_test)
print(accuracy)
```

```python
# %87'lik doğru tahmin.Şimdi tahminlere bakalım.

predictions = linear_regression.predict(X_test)
for i in range(len(predictions)):
    print(predictions[x], X_test[x], [y_test[x]])
```
